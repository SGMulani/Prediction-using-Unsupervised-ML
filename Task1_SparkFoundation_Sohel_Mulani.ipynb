```python
# Load the Iris dataset
data = load_iris()
X = data.data
```

```python
# Standardize the data (optional but recommended for hierarchical clustering)
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# Calculate the linkage matrix
linkage_matrix = linkage(X_std, method='ward', metric='euclidean')
```

```python
# Plot the dendrogram
plt.figure(figsize=(12, 6))
dendrogram(linkage_matrix, orientation="top")
plt.title('Dendrogram')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python

# Use the dendrogram to determine the optimal number of clusters
distances = pdist(X_std)
max_d = 2.8  # Adjust this value based on the dendrogram
clusters = fcluster(linkage_matrix, max_d, criterion='distance')
```

```python

# Visualize the clusters
plt.scatter(X_std[:, 0], X_std[:, 1], c=clusters, cmap='rainbow')
plt.title('Hierarchical Clustering')
plt.xlabel('Feature 1 (Standardized)')
plt.ylabel('Feature 2 (Standardized)')
plt.show()
```

```python
# Print the cluster assignments
for i in range(1, clusters.max() + 1):
    print(f'Cluster {i}:')
    print(data.target_names[data.target[clusters == i]])

# You can adjust 'max_d' based on the dendrogram to determine the number of clusters
```
